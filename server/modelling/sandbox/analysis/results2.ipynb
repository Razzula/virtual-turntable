```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('ResNet_Size_Time.csv')

# Fit a polynomial line of best fit (degree 2)
coeffs = np.polyfit(df["Image Size"], df["Time Taken (seconds)"], deg=2)
poly_fit = np.poly1d(coeffs)

# Generate smooth x values for the fitted curve
x_fit = np.linspace(df["Image Size"].min(), df["Image Size"].max(), 100)
y_fit = poly_fit(x_fit)

# Plot bar chart
plt.figure(figsize=(8, 6))
plt.bar(df["Image Size"], df["Time Taken (seconds)"], color="black", edgecolor="black", label="Time Taken")
plt.plot(x_fit, y_fit, color="red", linestyle="--", label="Best Fit Curve")

plt.xlabel("Image Size (pixels²)")
plt.ylabel("Time Taken per epoch (seconds)")
plt.title("Time Taken to Train on Different Image Sizes (~190 images per epoch)")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
